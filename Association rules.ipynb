```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
frequent_itemsets = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>1]
```

```python
plt.figure(figsize=(10,28))
sns.heatmap(book, annot = True)
plt.show()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
frequent_itemsets2=apriori(book,min_support=0.20,use_colnames=True)
frequent_itemsets2
```

```python
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.6)
rules2
```

```python
plt.figure(figsize=(12,10))
plt.scatter(rules2['support'],rules2['confidence'])
plt.show()
```

```python
frequent_itemsets_fp=fpgrowth(book, min_support=0.1, use_colnames=True,verbose=1)
print(frequent_itemsets_fp.shape)
```

```python
frequent_itemsets_fp.sort_values("support", ascending = False).head()
```

```python
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.5)
print(rules_fp.shape)
```

```python
rules_fp[(rules_fp.support > 0.15) & (rules_fp.confidence > 0.4)].sort_values("confidence", ascending = False).head()
```

```python
rules_fp['lhs items'] = rules_fp['antecedents'].apply(lambda x:len(x) )
rules_fp[rules_fp['lhs items']>1].sort_values('lift', ascending=False).head()
```

Interview Questions:
1.	What is lift and why is it important in Association rules?

      Lift is the confidence of the rule divided by expected confidence, assuming the itemsets x and y are independent of each other.A lift value greater than 1 indicates that the association between two items is stronger than expected based on the frequency of the individual items.

2.	What is support and Confidence. How do you calculate them?

         Support refers to  relative frequency of an item set in dataset. An item set can contain more than one item. The fruit item has apple, grape. Confidence is a reliability measure of a  association rule. confidence of A and B= support AUB/Support A

3.	What are challenges of Association rules mining?

            The metrics that measure the quality and relevance of the rules. The most commonly used metrics are support, confidence, and lift. Choosing the right metrics depends on the domain, the goal, and the data characteristics.
